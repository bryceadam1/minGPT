```python
import matplotlib.pyplot as plt
```

```python
import re

def extract_batch_and_loss(line):
    # Use regular expression to extract Batch and Loss information
    match = re.search(r'Batch: (\d+), Loss: ([\d.]+)', line)
    if match:
        batch_number = int(match.group(1))
        loss = float(match.group(2))
        return batch_number, loss
    return None

def process_log_file(file_path):
    batch_numbers = []
    losses = []
    with open(file_path, 'r') as file:
        for line in file:
            result = extract_batch_and_loss(line)
            if result:
                batch_number, loss = result
                batch_numbers.append(batch_number)
                losses.append(loss)
    
    return batch_numbers, losses
```

```python
batch, loss = process_log_file('log.txt')
plt.plot(batch, loss)
plt.xlabel('Batch')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.savefig('figures/loss.png')
plt.show()
```
